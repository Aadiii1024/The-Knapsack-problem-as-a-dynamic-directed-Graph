```python
import networkx as nx
import matplotlib.pyplot as plt

# Items: (weight, value)
items = [
    (3, 60),   # A
    (2, 100),  # B
    (4, 120),  # C
    (5, 150)   # D
]
item_names = ['A', 'B', 'C', 'D']
capacity = 10
n = len(items)

# DP table initialization
dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]

# Create graph
G = nx.DiGraph()

# Fill the DP table and build the graph
for i in range(1, n + 1):
    wt, val = items[i - 1]
    for w in range(capacity + 1):
        # Exclude current item
        dp[i][w] = dp[i - 1][w]
        G.add_edge(f"{i-1},{w}", f"{i},{w}", label="exclude")

        # Include current item if it fits
        if w >= wt:
            include_val = dp[i - 1][w - wt] + val
            if include_val > dp[i][w]:
                dp[i][w] = include_val
                G.add_edge(f"{i-1},{w-wt}", f"{i},{w}", label=f"include {item_names[i-1]}")

# Final result
print("Maximum value:", dp[n][capacity])

# Draw the graph (optional trimming for clarity)
pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(14, 10))

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=500, node_color='lightblue')
nx.draw_networkx_labels(G, pos, font_size=8)

# Draw edges with labels
edges = G.edges(data=True)
nx.draw_networkx_edges(G, pos, edgelist=edges, arrowstyle='->', arrowsize=10)
edge_labels = {(u, v): d['label'] for u, v, d in edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=6)

plt.title("Knapsack Problem as Dynamic Directed Graph")
plt.axis('off')
plt.show()
```
